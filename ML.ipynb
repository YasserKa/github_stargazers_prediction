```python
# Using a local dataset
import json
import pandas as pd

with open('repos1/gistfile1.txt') as json_file:
    repos = json.load(json_file)
```

```python
df_repos = pd.DataFrame.from_dict(repos).T #Transposing the dataframe column=>row, row=>column
```

```python
df_repos.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#Random Forest
X = df_repos.drop(['owner', 'name', 'id', 'full_name', 'created_at', 'last_commit', 'stargazer_count'], axis=1)
y = df_repos['stargazer_count']
y = y.astype('int')

#sc = StandardScaler()
#X = sc.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn import metrics

rf = RandomForestRegressor(n_estimators=200, random_state = 42)
rf.fit(X_train, y_train)

training_score = rf.score(X_train, y_train)
test_score = rf.score(X_test, y_test)

print("training score: ", training_score)
print("test score: ", test_score)
```
